```python
import numpy as np
import pandas as pd
```

```python
card_data=pd.read_csv('creditcard.csv')
```

```python
card_data.head()
```

```python
card_data.tail()
```

```python
#distribution of two classes
card_data['Class'].value_counts()
```

```python
#0 is legit and 1 is fraud
legit=card_data[card_data.Class==0]
fraud=card_data[card_data.Class==1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
legit_sample=legit.sample(n=25)
```

```python
print(legit_sample.shape)
```

```python
#concate the dataframe
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```
